```python
# ==========================================
# FEATURE ENGINEERING: RFM & CLV
# ==========================================
# === RFM ANALYSIS ===
# Your dataset already has these features:
# Recency  = DaySinceLastOrder (lower = more recent = better)
# Frequency = OrderCount (higher = more orders = better)
# Monetary  = CashbackAmount (proxy for spending — higher = better)
# === IMPORT LIBRARIES ===
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# Ensure output directories exist
os.makedirs("outputs", exist_ok=True)
os.makedirs("plots", exist_ok=True)
```

```python
# === CREATE RFM DATASET ===
rfm = df.groupby('CustomerID').agg({
    'DaySinceLastOrder': 'min',   # Recency: Lower is better
    'OrderCount': 'sum',          # Frequency: Higher is better
    'CashbackAmount': 'sum'       # Monetary: Higher is better
}).reset_index()

# Rename columns
rfm.rename(columns={
    'DaySinceLastOrder': 'Recency',
    'OrderCount': 'Frequency',
    'CashbackAmount': 'Monetary'
}, inplace=True)

print("RFM Dataset Preview:")
print(rfm.head())
```

```python
# === CALCULATE RFM SCORES ===

# Recency: Lower is better
rfm['R_Score'] = pd.qcut(
    rfm['Recency'],
    q=5,
    labels=[5, 4, 3, 2, 1],
    duplicates='drop'
)

# Frequency: Higher is better
rfm['F_Score'] = pd.qcut(
    rfm['Frequency'].rank(method='first'),
    q=5,
    labels=[1, 2, 3, 4, 5],
    duplicates='drop'
)

# Monetary: Higher is better
rfm['M_Score'] = pd.qcut(
    rfm['Monetary'].rank(method='first'),
    q=5,
    labels=[1, 2, 3, 4, 5],
    duplicates='drop'
)

# Convert scores to integers
rfm[['R_Score', 'F_Score', 'M_Score']] = rfm[
    ['R_Score', 'F_Score', 'M_Score']
].astype(int)

# Compute total RFM score
rfm['RFM_Score'] = (
    rfm['R_Score'] +
    rfm['F_Score'] +
    rfm['M_Score']
)

print("\nRFM Scores Preview:")
print(rfm.head())
```

```python
# Create customer segments
def segment_customer(row):
    r, f, m = row['R_Score'], row['F_Score'], row['M_Score']
    
    if r >= 4 and f >= 4 and m >= 4:
        return 'Champions'
    elif f >= 3 and m >= 3:
        return 'Loyal'
    elif r >= 4 and f <= 2:
        return 'New Customers'
    elif r <= 2 and f >= 3:
        return 'At Risk'
    elif r <= 2 and f <= 2:
        return 'Lost'
    else:
        return 'Needs Attention'

rfm['Segment'] = rfm.apply(segment_customer, axis=1)

```

```python
# === VISUALIZE RFM SEGMENTS ===
plt.figure(figsize=(10, 6))
rfm['Segment'].value_counts().plot(kind='bar')
plt.title('Customer Segments (RFM Analysis)')
plt.xlabel('Segment')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('plots/rfm_segments.png')
plt.show()
```

```python
# === MERGE WITH ORIGINAL DATA ===
# Remove existing RFM-related columns if re-running the notebook
cols_to_drop = [
    'Recency', 'Frequency', 'Monetary',
    'R_Score', 'F_Score', 'M_Score',
    'RFM_Score', 'Segment'
]
df = df.drop(columns=[col for col in cols_to_drop if col in df.columns])

# Merge RFM metrics and segments into the original dataset
df = df.merge(
    rfm[
        [
            'CustomerID',
            'Recency', 'Frequency', 'Monetary',
            'R_Score', 'F_Score', 'M_Score',
            'RFM_Score', 'Segment'
        ]
    ],
    on='CustomerID',
    how='left'
)

print("Merged Dataset Preview:")
print(df[['CustomerID', 'R_Score', 'F_Score', 'M_Score', 'Segment']].head())
```

```python
# === CHURN RATE BY SEGMENT ===
churn_by_segment = (
    df.groupby('Segment')['Churn']
    .agg(['count', 'sum', 'mean'])
    .rename(columns={
        'count': 'Total_Customers',
        'sum': 'Churned_Customers',
        'mean': 'Churn_Rate'
    })
    .sort_values(by='Churn_Rate', ascending=False)
)

# Convert churn rate to percentage
churn_by_segment['Churn_Rate (%)'] = (
    churn_by_segment['Churn_Rate'] * 100
).round(2)

print("\nChurn Rate by Segment:")
print(churn_by_segment)
```

```python
import matplotlib.pyplot as plt

churn_by_segment['Churn_Rate (%)'].sort_values().plot(
    kind='barh',
    figsize=(10, 6),
    title='Churn Rate by Customer Segment (%)'
)

plt.xlabel('Churn Rate (%)')
plt.ylabel('Segment')
plt.tight_layout()
plt.savefig('outputs/churn_by_segment.png')
plt.show()
```

```python
# Check: what does high Recency look like?
print(rfm[['Recency', 'R_Score']].head(20))
```

```python
# === CUSTOMER LIFETIME VALUE (CLV) ===
df['CLV_Estimate'] = df['CashbackAmount'] * df['OrderCount'] * (df['Tenure'] / 12)

# CLV by segment
clv_by_segment = df.groupby('Segment')['CLV_Estimate'].mean().sort_values(ascending=False)
print("\nAverage CLV by Segment:")
print(clv_by_segment)

# KEY CHART: Churn Risk vs Customer Value
segment_summary = df.groupby('Segment').agg(
    churn_rate=('Churn', 'mean'),
    avg_clv=('CLV_Estimate', 'mean'),
    count=('CustomerID', 'count')
).reset_index()

# Convert churn rate to percentage
segment_summary['churn_rate_pct'] = segment_summary['churn_rate'] * 100

# === VISUALIZATION: CHURN RISK VS CLV ===
plt.figure(figsize=(10, 7))
plt.scatter(
    segment_summary['churn_rate_pct'],
    segment_summary['avg_clv'],
    s=segment_summary['count'] * 2,
    alpha=0.7
)

# Annotate segments
for _, row in segment_summary.iterrows():
    plt.annotate(
        row['Segment'],
        (row['churn_rate_pct'], row['avg_clv']),
        fontsize=10,
        ha='center',
        va='bottom'
    )

plt.xlabel('Churn Rate (%)')
plt.ylabel('Average Customer Lifetime Value')
plt.title('Customer Segments: Churn Risk vs Value')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('plots/churn_vs_clv.png', dpi=300)
plt.show()

# === SAVE OUTPUTS ===
segment_summary.to_csv("outputs/segment_summary.csv", index=False)
clv_by_segment.to_csv("outputs/clv_by_segment.csv")
```

```python
# Retraining models after RFM analysis for improvements.
# Merge RFM features into encoded dataset
rfm_features = rfm[
    ['CustomerID', 'Recency', 'Frequency', 'Monetary',
     'R_Score', 'F_Score', 'M_Score', 'RFM_Score', 'Segment']
]

df_enhanced = df_encoded.merge(rfm_features, on='CustomerID', how='left')

# One-hot encode Segment
df_enhanced = pd.get_dummies(df_enhanced, columns=['Segment'], drop_first=True)

# Ensure CLV is included
df_enhanced['CLV_Estimate'] = df['CLV_Estimate']

# Drop identifier
df_enhanced.drop(columns=['CustomerID'], inplace=True)
df_enhanced.info()
```

```python
#Splitting the target and feature variables
from sklearn.model_selection import train_test_split

X_enhanced = df_enhanced.drop(columns=['Churn'])
y_enhanced = df_enhanced['Churn']

X_train_e, X_test_e, y_train_e, y_test_e = train_test_split(
    X_enhanced,
    y_enhanced,
    test_size=0.2,
    random_state=42,
    stratify=y_enhanced
)
```

```python
#feature scaling scaling 
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled_e = scaler.fit_transform(X_train_e)
X_test_scaled_e = scaler.transform(X_test_e)
```

```python
#Linear regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score
log_model_e = LogisticRegression(max_iter=1000, random_state=42)
log_model_e.fit(X_train_scaled_e, y_train_e)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model_e = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,
    min_samples_split=5,
    random_state=42
)


rf_model_e.fit(X_train_e, y_train_e)
print("Train accuracy:", rf_model_e.score(X_train_e, y_train_e))
print("Test accuracy:", rf_model_e.score(X_test_e, y_test_e))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report

gbc_model_e = GradientBoostingClassifier(
    n_estimators=100, 
    random_state=42
)
gbc_model_e.fit(X_train_e, y_train_e)
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score, f1_score
import pandas as pd

def evaluate_model(name, model, X_test, y_test):
    y_pred = model.predict(X_test)
    
    # probability handling (works for sklearn models)
    y_prob = model.predict_proba(X_test)[:, 1]
    
    return {
        "Model": name,
        "Accuracy": accuracy_score(y_test, y_pred),
        "F1": f1_score(y_test, y_pred),
        "ROC-AUC": roc_auc_score(y_test, y_prob)
    }

results_enhanced = [
    evaluate_model("Logistic Regression (Enhanced)", log_model_e, X_test_scaled_e, y_test_e),
    evaluate_model("Random Forest (Enhanced)", rf_model_e, X_test_e, y_test_e),
    evaluate_model("gradient boost (Enhanced)", gbc_model_e, X_test_scaled_e, y_test_e)
]

df_results = pd.DataFrame(results_enhanced)

# optional: nicer formatting
df_results["Accuracy"] = df_results["Accuracy"].round(3)
df_results["F1"] = df_results["F1"].round(3)
df_results["ROC-AUC"] = df_results["ROC-AUC"].round(3)

df_results
```
